```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set(font='Lato')
```

# Alucar - Analisando as vendas

```python
alucar = pd.read_csv('data/alucar.csv')
alucar.head()
```

```python
print('Size:\n',alucar.shape)
print('NaN Values:\n',alucar.isna().sum())
print('Types:\n', alucar.dtypes)
```

```python
# cast month column to datetime
alucar.mes = pd.to_datetime(alucar.mes)
alucar.dtypes
```

```python
ax = sns.lineplot(x='mes',y='vendas',data=alucar)
ax.figure.set_size_inches(12,6)
ax.set_title('Vendas Alucar 2017/2018', loc='left', fontsize=18)
ax.set_xlabel('Tempo', fontsize=14)
ax.set_ylabel('Vendas (R$)', fontsize=14)
ax = ax
```

```python
alucar['aumento'] = alucar['vendas'].diff()
alucar.head()
```

```python
def plotar(title, labelx, labely, x, y, dataset):
    ax = sns.lineplot(x=x, y=y,data=dataset)
    ax.figure.set_size_inches(12,6)
    ax.set_title(title, loc='left', fontsize=18)
    ax.set_xlabel(labelx, fontsize=14)
    ax.set_ylabel(labely, fontsize=14)
    ax = ax
```

```python
plotar('Aumento das vendas 2017/2018',
        'Tempo','Aumento','mes','aumento',alucar)
```

```python
alucar['aceleracao'] = alucar['aumento'].diff()
alucar.head()
```

```python
plotar('Aceleracao das vendas 2017/2018',
        'Tempo','Aumento','mes','aceleracao',alucar)
```
